```python
# Import the dependencies.
import pandas as pd
import numpy as np
```

```python
#  Read in the train dataset. 
train_df = pd.read_csv("./Resources/train.csv")
train_df.head()
```

```python
#  Read in the test dataset. 
test_df = pd.read_csv("./Resources/test.csv")
test_df.head()
```

```python
# Get a brief summary of the train DataFrame.
train_df.info()
```

```python
# Get a brief summary of the test DataFrame.
test_df.info()
```

```python
train_df.columns
```

```python
test_df.columns
```

```python
# Adding the test_df to the train_df
added_df = pd.concat([train_df, test_df])
added_df
```

```python
added_df.columns
```

```python
# Changing the column names
combined_df = added_df.rename(columns = {'BusinessTravel' :  'Travel Frequency' , 'DailyRate' : 'Daily Rate',
       'DistanceFromHome' : 'Distance from Home', 'Education' : 'Education Level' , 'EducationField' : 'Educational Field',
       'EmployeeNumber' : 'Employee #', 'EnvironmentSatisfaction' : 'Environmental Satisfaction Rating', 'HourlyRate' : 'Hourly Rate',
       'JobInvolvement': 'Involvement Rating', 'JobLevel': 'Level', 'JobRole': 'Position', 'JobSatisfaction':'Job Satisfaction Rating',
       'MaritalStatus': 'Marital Status', 'MonthlyIncome':'Monthly Income', 'MonthlyRate' : 'Monthly Pay Rate', 'NumCompaniesWorked': '# Companies Worked',
       'OverTime':'Over Time Employee', 'PercentSalaryHike':'Percent Salary Increase', 'PerformanceRating' :'Performance Rating',
       'RelationshipSatisfaction':'Relationship Satisfaction Rating', 'StandardHours': 'Standard Hours', 'StockOptionLevel' :'Stock Option Level',
       'TotalWorkingYears':'Total Working Years', 'TrainingTimesLastYear':'# Training sessions last year' , 'WorkLifeBalance':'Work Life Balance Level',
       'YearsAtCompany': 'Years at Company', 'YearsInCurrentRole':'Years in Current Role', 'YearsSinceLastPromotion': 'Years since last promotion',
       'YearsWithCurrManager': 'Years with Current Manager'})
combined_df.head()
```

```python
# Filling values with NA
combined_df = combined_df.fillna('Na')
combined_df
```

```python
# Export to csv file
combined_df.to_csv("Resources/combined_dataset.csv", index=False)
```

```python
# Get a brief summary of the merged DataFrame.
combined_df.info()
```
